```python
casas_baratas= casas.sort_values('price',ascending= False).iloc[300:]
```

```python
casas_baratas.head()
```

```python
plt.figure(figsize= (10,8))
sb.scatterplot (x= 'long',y='lat',data= casas_baratas, hue= 'price',edgecolor= None, alpha=0.5)
plt.show()
```

```python
casas_baratas['date']=pd.to_datetime(casas_baratas['date'])
```

```python
casas_baratas['year']= casas_baratas['date'].apply(lambda date:  date.year)
```

```python
casas_baratas['month']= casas_baratas['date'].apply (lambda date: date.month)
```

```python
casas_baratas=casas_baratas.drop(['id','date'], axis=1)
```

```python
casas_baratas.head()
```

```python
X = casas_baratas.drop('price',axis=1).values
y = casas_baratas['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
# scaler.fit (X_train)
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test= scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
# 20 variables,pues 20 neuronas
```

```python
model= Sequential()
```

```python
model.add( Dense(20,activation='relu'))
model.add( Dense(20,activation='relu'))
model.add( Dense(20,activation='relu'))
model.add( Dense(20,activation='relu'))

model.add( Dense(1))
# modelo de regresiÃ³n:
model.compile (optimizer='adam', loss= "mse")
```

```python
model.fit (x= X_train, y=y_train, validation_data=(X_test,y_test),batch_size=128, epochs= 400)
# batchsize= parte = 2,64,128,256... el nÂº mas bajo tarda mas pero evita overfitting
```

```python
losses= pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions= model.predict (X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)**0.5
```

```python
mean_absolute_error(y_test,predictions)
```

```python
explained_variance_score(y_test,predictions)
```
