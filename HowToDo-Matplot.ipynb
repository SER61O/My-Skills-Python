# Matplotlib

```python
#Importar siempre es lo primero 
import matplotlib.pyplot as plt
```

```python
año=[1960,1970,1980,1990,2000,2010,2020]
población= [1.8,2.4,2.9,3.2,3.9,4.7,5.9]
```

```python
#con plot se muestra grafico de linea
plt.plot (año,población)
plt.show()
```

```python
#con scatter se muestra grafico de puntos
#se puede meter plt.show() en la misma linea de código
plt.show(plt.scatter(año,población))
```

## Histograma

```python
?plt.hist
```

```python
valores=[1,2,3,3,5,6,7,7]
plt.hist(valores)
```

```python
#para que no quede tan poco estético es mejor unir
#se pueden meter muchas modificaciones que se ven en ?.plt.hist como color..borde..
plt.hist(valores, bins=4, color="yellow", edgecolor="black")
plt.show()
```

## Etiquetas

```python
# XY label= titulos ejes
#title= titulo
#XY ticks = numeros ejes
plt.hist(valores, bins=4, color="yellow", edgecolor="black")
plt.ylabel ("Población")
plt.xlabel ("Año")
plt.title ("Crecimiento")
plt.yticks([0,1,2,3,4])
plt.show()
```
