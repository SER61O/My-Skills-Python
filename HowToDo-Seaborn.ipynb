```python
import seaborn as sb
```

```python
sb.get_dataset_names()
```

```python
iris= sb.load_dataset('iris')
```

```python
iris.head()
```

```python
import matplotlib.pyplot as plt
```

```python
indices= iris[iris['petal_length'] >= 2.5].index
iris.loc[indices,"Color"] ="r"
indices= iris[iris['petal_length'] < 2.5].index
iris.loc[indices,"Color"] = "b"
plt.scatter(iris['petal_length'],iris['petal_width'], c= iris.loc[:,"Color"])
plt.title('Dimensiones de los petalos')
plt.show()
```

# Como hacemos esto en SEABORN

```python
sb.lmplot(x='petal_length',y='petal_width', data= iris)
plt.show()
```

```python
# quitar la regresión con fit_reg=0
#clasificar con hue ( aqui colores)
# con palette cambiamos colores
sb.lmplot(x='petal_length',y='petal_width', data= iris, fit_reg=0, hue='species', palette='Set1')
plt.show()
```

```python
# podemos hacer 1 grafico para cada una con Row
sb.lmplot(x='petal_length',y='petal_width', data= iris, fit_reg=0, hue='species', palette='Set1', row='species')
plt.show()
```

## otros tipos de gráficos 

```python
# facil, copiar y cambiar
# ?tipodegrafico para ver los parametros de cada uno,facil todo lo que se puede modificar
```

```python
#boxplot
sb.boxplot(x='species',y='petal_length', data= iris)
plt.show()
```

```python
#strip 
sb.stripplot(x='species',y='petal_length', data= iris)
plt.show()
```

```python
#swarmplot
sb.swarmplot(x='species',y='petal_length', data= iris)
plt.show()
#el error que produce es que el 14% de los puntos no tienen espacio para colocarse, habria que cambiar el tamaño con size
```

```python
# grafico de barras
# con dataset titanic
titanic= sb.load_dataset('titanic')
titanic.head()
```

```python
sb.barplot(x='sex',y='survived',hue='class',data=titanic)
plt.show()
```

```python
#cuenta con countplot
#con palet podemos cambiar colores...
sb.countplot(x='deck',data=titanic, palette="Greens_d")
plt.show()
```

```python
#pointplot
#mas formas de cambiar colores con palette{}....
# marcadores con markers[]
#cambiar linea con linestyle
sb.pointplot(x='class',y='survived',hue='sex',data=titanic, palette={'male':'g','female':'m'}, markers=['^','o'], linestyles=["-","--"])
plt.show()
```

```python
#violinplot
sb.violinplot(x='age',y='sex',hue='survived',data=titanic)
plt.show()
```

# Gráficos especiales 

```python
# se pueden juntar 2 tipos de graficos
# 1º
# inner=(linea del medio del grafico anterior)
sb.violinplot(x='age',y='sex',data=titanic, inner= None, color='lightgray')
plt.show()
```

```python
#2º
sb.violinplot(x='age',y='sex',data=titanic, inner= None, color='lightgray')
#podemos cambiar el tamaño del segundo grafico solo
sb.swarmplot(x='age',y='sex',data=titanic, size=3)
plt.title('Edad de hombres y mujeres')
plt.show()
```
